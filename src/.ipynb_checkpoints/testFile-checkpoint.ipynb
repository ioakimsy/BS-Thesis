# Graphing and Benchmarking

## For loops vs function.

```julia
@time begin
lower_bound = 0
higher_bound = 8*pi
elements = 1000
x = LinRange(lower_bound, higher_bound, elements);
y = [];
for index in range(1,elements)
    push!(y,sin(x[index]));
end
end

@time begin
    lower_bound = 0
    higher_bound = 8*pi
    elements = 1000
    x = LinRange(lower_bound, higher_bound, elements);
    y = sin.(x);
end
```

## Variables can store plots!

```julia
sinPlot = plot(x,y)
```

## Overlapping Plots

```julia
@time begin
lower_bound = 0
higher_bound = 8*pi
elements = 1000
x = LinRange(lower_bound, higher_bound, elements)
y = sin.(x)
z = cos.(x)
plot(x,y, label="sin(x)", minorgrid=true)
plot!(x,z, label="cos(x)")
end
```

# Symbolics

## Making variables

```julia
@variables n m
j = n^2 + m^3
```

## Differential Operators

```julia
Symbolics.derivative(j,m)
```

```julia
D_m = Differential(m)
D_m(j)
```

```julia
#perform the derivative
k = expand_derivatives(D_m(j))
```

```julia
#find which variable we are differentiating with respect to
D_m.x
```

```julia
V = substitute.(k, (Dict(m => 2.0),))
```

# Edison Codes

```julia
@benchmark begin
array = [5,6,7,7,8,8]
additions = 0
max_additions = 4

while (additions<max_additions)
    #print(array)
    sort(array)
    append!(array,array[1]+array[2])
    popfirst!(array)
    popfirst!(array)
    sort(array)
    additions += 1
end

maximum(array)
end
```
