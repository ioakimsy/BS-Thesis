```python
def fib(n):
    output = [0,1]
    if n<1:
        return "Error"
    elif n==1:
        return output[0]
    elif n == 2:
        return output[1]

    else:
        while len(output)<n:
            new_fib = output[-1] + output[-2]
            output.append(new_fib)
        return output
```

```python
fib(10)
```

```python
print(range(10))
```

```python
y = fib(100)
x = [element+1 for element in range(len(y))]
```

```python
plt.plot(x,y)
plt.xlabel("term")
plt.ylabel("value of the nth term of the fibonacci sequence")
```

```python
def binary(n):
    exp=0
    y=[]
    while n > (2**exp):
        exp+=1
    exp = exp-1
    while n > 0:
        if n >= (2**exp):
            y.append(1)
            n=n%(2**exp)
        else:
            y.append(0)
        exp -= 1
        print(n)
    print(y)
    
```
