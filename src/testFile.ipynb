# Graphing and Benchmarking

## For loops vs function.

```julia
@time begin
lower_bound = 0
higher_bound = 8*pi
elements = 1000
x = LinRange(lower_bound, higher_bound, elements);
y = [];
for index in range(1,elements)
    push!(y,sin(x[index]));
end
end

@time begin
    lower_bound = 0
    higher_bound = 8*pi
    elements = 1000
    x = LinRange(lower_bound, higher_bound, elements);
    y = sin.(x);
end
```

## Variables can store plots!

```julia
sinPlot = plot(x,y)
```

## Overlapping Plots

```julia
@time begin
lower_bound = 0
higher_bound = 8*pi
elements = 1000
x = LinRange(lower_bound, higher_bound, elements)
y = sin.(x)
z = cos.(x)
plot(x,y, label="sin(x)", minorgrid=true)
plot!(x,z, label="cos(x)")
end
```

# Edison Codes

```julia
@benchmark begin
array = [5,6,7,7,8,8]
additions = 0
max_additions = 4

while (additions<max_additions)
    #print(array)
    sort(array)
    append!(array,array[1]+array[2])
    popfirst!(array)
    popfirst!(array)
    sort(array)
    additions += 1
end

maximum(array)
end
```

```julia
x_hist_vals = 1:12
y_hist_vals = [2,4,11,12,16,10,10,11,7,5,3,2]
y_hist_vals_normalized = y_hist_vals / sum(y_hist_vals)
ave = sum(x_hist_vals .* y_hist_vals) / sum(y_hist_vals)
R = ave * sqrt(2/π)

marble_data = []
for i in 1:12
    for j in 1:y_hist_vals[i]
    push!(marble_data,x_hist_vals[i])
    end
end

r = 0.1:0.1:12
f_r = (r ./ R^2) .* exp.(-(r.^2) ./ (2*R^2))
normalized_f_r = f_r ./ sum(f_r)
```

```julia
histogram(marble_data, normalize=:true, bins=range(1,12,length=12),label="Experimental data",dpi=300,xticks=1:12)
plot!(r.+1,f_r, label="Theoretical distribution",linewidth=2)
title!("Number of hits per ring")
ylabel!("Number of hits (normalized)")
xlabel!("Ring number")
#savefig("191 A2 toss histogram")
```

```julia
num_heads = 0:10
f_1 = [0,1,3,12,17,31,16,12,7,1,0]
f_2 = [0,0,4,16,18,22,21,11,7,0,1]
fⱼˢᵗ = []
for i in num_heads
    push!(fⱼˢᵗ, factorial(10) / (factorial(10 - num_heads[i+1]) * factorial(num_heads[i+1]) * 2^10))
end
trial_1 = []
for i in 1:11
    for j in 1:f_1[i]
        push!(trial_1,num_heads[i])
    end
end
trial_2 = []
for i in 1:11
    for j in 1:f_2[i]
        push!(trial_2,num_heads[i])
    end
end

combined_trials = vcat(trial_1,trial_2)
```

```julia
histogram(trial_1, normalize=:true, label = "trial 1", legend=:topleft, 
title = "Histogram for number of heads per 10 coins tossed 100 times - Trial 1",
titlefontsize=10, xticks=0:10,
dpi=300)
plot!(0.5:10.5,fⱼˢᵗ, label = "theoretical distribution", linewidth=2)
xlabel!("Number of heads per 10 coins")
ylabel!("Frequency (Normalized)")
xlims!(0,11)
savefig("191 A2 Coin toss trial 1")
```

```julia
histogram(trial_2, normalize=:true, label = "trial 2", legend=:topleft,
title = "Histogram for number of heads per 10 coins tossed 100 times - Trial 2",
titlefontsize=10, xticks=0:10,
dpi=300)
plot!(0.5:10.5,fⱼˢᵗ, label = "theoretical distribution", linewidth=2)
xlabel!("Number of heads per 10 coins")
ylabel!("Frequency (Normalized)")
xlims!(0,11)
savefig("191 A2 Coin toss trial 2")
```
